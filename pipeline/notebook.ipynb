```python
dataType = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

dateColumns = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    url,
    dtype=dataType,
    parse_dates=dateColumns
)
```

```python
# create_engine builds the connection configuration to PostgreSQL
# text allows us to write raw SQL strings (used for TRUNCATE later)
from sqlalchemy import create_engine, text

# Define WHERE and HOW to connect to PostgreSQL
# format: postgresql://username:password@host:port/database_name
# does NOT open a connection yet — just stores the credentials
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
#creates the table in PostgreSQL but inserts zero rows
df.head(0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
#This just prints the SQL CREATE TABLE statement to your screen so you can see it. 
# It doesn't do anything to the database — it's purely for inspection.
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
# Import progress bar library — .auto detects notebook vs terminal automatically
from tqdm.auto import tqdm

# Recreate iterator fresh to avoid exhausted iterator bug
df_iter = pd.read_csv(
    url,
    dtype=dataType,
    parse_dates=dateColumns,
    chunksize=100000,
    iterator=True
)

# Pull first chunk to create table structure
first_chunk = next(df_iter)

# Create table structure only — no rows inserted yet
# replace means: drop and recreate the table if it exists
first_chunk.head(0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
print("Table created")

# Insert first chunk's actual data
first_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
print("Inserted first chunk:", len(first_chunk))

# Loop through all remaining chunks and insert them
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
    print("Inserted chunk:", len(df_chunk))

print("Done! All data inserted into yellow_taxi_data.")
```
